## Use ordinalencoder to turn the ordinal columns to numerical

- electrical
- exter_qual
- bsmt_qual
- bsmtfin_type_1
- garage_type

> electrical

```python
electrical_cate=['Mix','FuseP','FuseA','FuseF','SBrkr']
oe=OrdinalEncoder(categories=[electrical_cate],dtype=int)
test_df['electrical']=oe.fit_transform(test_df[['electrical']])
```

> exter_qual

```python
exter_qual_cate=['Po','Fa','TA','Gd','Ex']
oe=OrdinalEncoder(categories=[exter_qual_cate],dtype=int)
test_df['exter_qual']=oe.fit_transform(test_df[['exter_qual']])
```

> bsmt_qual

```python
bsmt_qual_cate=['NA','Po','Fa','TA','Gd','Ex']
oe=OrdinalEncoder(categories=[bsmt_qual_cate],dtype=int)
test_df['bsmt_qual']=oe.fit_transform(test_df[['bsmt_qual']])
```

> bsmtfin_type_1

```python
bsmtfin_type_1_cate=['NA','Unf','LwQ','Rec','BLQ','ALQ','GLQ']
oe=OrdinalEncoder(categories=[bsmtfin_type_1_cate],dtype=int)
test_df['bsmtfin_type_1']=oe.fit_transform(test_df[['bsmtfin_type_1']])
```

>garage_type

```python
garage_type_cate=['NA','Detchd','CarPort','BuiltIn','Basment','Attchd','2Types']
oe=OrdinalEncoder(categories=[garage_type_cate],dtype=int)
test_df['garage_type']=oe.fit_transform(test_df[['garage_type']])
```

## Match the columns in test_df with train_df

```python
train_df=pd.read_csv('./datasets/improved_model_clean_train_data.csv')
```
